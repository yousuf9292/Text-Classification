```python
import pandas as pd 
```

```python
df=pd.read_csv('C:\\Users\\yousuf\\Downloads\\imdb-dataset-of-50k-movie-reviews\\IMDB Dataset.csv')
```

```python
from tensorflow import keras
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
df.head()
```

```python
X=df['review']
```

```python
y=df['sentiment']
```

```python
len(X)
```

```python
len(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_sentence, test_sentence, train_labels, test_labels=train_test_split(X,y,stratify=y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Le=LabelEncoder()
```

```python
Le.fit(train_labels)
```

```python
encoded_train_label=Le.transform(train_labels)
encoded_test_label=Le.transform(test_labels)
```

```python
encoded_train_label.shape
```

```python
train_sentence.shape
```

```python
max_features=1000
```

```python
tokenizer=Tokenizer(num_words=max_features,oov_token="<OOV>")
```

```python
tokenizer.fit_on_texts(train_sentence)
```

```python
word_ind=tokenizer.word_index
```

```python
print(word_ind)
```

```python
training_sequences=tokenizer.texts_to_sequences(train_sentence)
training_padded=pad_sequences(training_sequences)
```

```python
testing_sequences=tokenizer.texts_to_sequences(test_sentence)
testing_padded=pad_sequences(testing_sequences)
```

```python
print(training_padded[0])
```

```python
training_padded.shape
```

```python
from keras.models import Sequential
from keras.layers import Embedding,Dense,GlobalAveragePooling1D
```

```python
model=Sequential()
model.add(Embedding(max_features,128))
model.add(GlobalAveragePooling1D())
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
num_epochs=10
model.fit(training_padded,encoded_train_label,validation_data=(testing_padded,encoded_test_label),epochs=num_epochs)
```

```python
sample_json={
    "Text":["The movie hobit was so good i love it but the ending part was not to my expectation"]
}
```

```python
def return_prediction(model,tokenizer,sample_json):
    sentence=sample_json['Text']
    sequence=tokenizer.texts_to_sequences(sentence)
    padded=pad_sequences(sequence)
    result=model.predict(padded)
    return result
```

```python
return_prediction(model,tokenizer,sample_json)
```
